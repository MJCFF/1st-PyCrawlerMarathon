```python
import json
import requests
```

```python
r = requests.get('https://api.github.com/events')
r.text
```

```python
json.loads(r.text)
```

```python
print(type(r.text),type(json.loads(r.text)))
```

r.text用str型態儲存資料,json.loads則是用list方式儲存資料
